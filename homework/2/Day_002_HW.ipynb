## 客服機器人

### 理由
1. 資料蒐集便利(客服通常都已有明確的正確答案)
2. 容易制定SOP(客服通常有明確的通話流程)
3. 可透過問答的方式引導對話，使對話內容不會過於發散
4. 可由後台統計最常發問的問題，並加強該對話情境的相關詞彙，以增加用戶體驗，及回答精準度
5. 前兩項有利於監督式學習，讓電腦能更準確判斷用戶意圖
6. 客服比較不會遇到反諷，雙關的問題(比起各大網路平台的留言區)，更容易精準判斷前後文關係。

### 缺點
1. 目前雖然已有大量的客服機器人，尤其銀行業，但多數的機器人很容易被問倒，也常常答非所問，最終仍需導向真人客服。
2. 針對上述問題，我認為最大的難題是精確掌握發問方的發問意圖，一旦認錯問題，回答也必定是錯誤的，且中文的斷詞非常容易斷錯，造成後續文字處理出錯

### 期待的解決方法
1. 部分中文的字，本身就有其意義，我覺得從根本上讓機器了解單字意義(如同人類學習第一母語那樣)比無意義的斷詞或許會更有用一些

### 難處
1. 文字是富含情感的表達方式，然而電腦對於富有情感的抽象概念較難精準判斷
2. 文字量及單字量相當巨大(雖然最常用的部分可能沒有想像中的多)
3. 時事用語隨著時間時常改變，引用、轉注、假借的狀況數見不鮮
